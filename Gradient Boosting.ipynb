### Verinin alınması ve grafk gösterimi

```python
import numpy as np
import pandas as pd

data = pd.read_excel('data.xlsx',sheet_name='Sheet4')

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

fig, ax = plt.subplots(nrows=1, ncols=1,figsize=(8,6))
sns.scatterplot(x="x", y="hedef", data=data, cmap='coolwarm', s=100).set_title('örnek küme')
fig.savefig("gb_00.png", dpi=500)
```

### 1. iterasyon

```python
#model
from sklearn.ensemble import GradientBoostingRegressor

x_train = data[['x','y']]
y_train = data['hedef']

model = GradientBoostingRegressor(n_estimators=1)
model.fit(x_train,y_train)
y_pred = model.predict(x_train)

data['tahmin'] = y_pred
data['kalan'] = data['hedef'] - data['tahmin']


#grafik
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(16,6))

sns.scatterplot(x="x", y="hedef", data=data, cmap='coolwarm', s=100,
                ax=ax[0]).set_title('örnek küme ve tahmin (1. iterasyon)', fontsize=20)
sns.lineplot(x="x", y="tahmin", data=data, color='red', alpha=0.8, ax=ax[0])
ax[0].set_ylim(0,100)
ax[0].set(ylabel='hedef ve tahmin')

sns.scatterplot(x="x", y="kalan", data=data, cmap='coolwarm', s=100,
                ax=ax[1]).set_title('tahmin ile hedef arasındaki fark (1. iterasyon)', fontsize=20)
sns.lineplot(x="x", y=0, data=data, color='red', alpha=0.5, ax=ax[1])
ax[1].set_ylim(-50,50)

fig.savefig("gb_01.png", dpi=500)
```
