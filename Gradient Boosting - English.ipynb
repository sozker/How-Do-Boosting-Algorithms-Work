### Verinin alınması ve grafk gösterimi

```python
import numpy as np
import pandas as pd

data = pd.read_excel('data.xlsx',sheet_name='Sheet5')

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

fig, ax = plt.subplots(nrows=1, ncols=1,figsize=(8,6))
sns.scatterplot(x="x", y="target", data=data, cmap='coolwarm', s=100).set_title('dataset')
fig.savefig("gb_00_eng.png", dpi=500)
```

### 1. iterasyon

```python
#model
from sklearn.ensemble import GradientBoostingRegressor

x_train = data[['x','y']]
y_train = data['target']

model = GradientBoostingRegressor(n_estimators=1)
model.fit(x_train,y_train)
y_pred = model.predict(x_train)

data['prediction'] = y_pred
data['residual'] = data['target'] - data['prediction']


#grafik
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(16,6))

sns.scatterplot(x="x", y="target", data=data, cmap='coolwarm', s=100,
                ax=ax[0]).set_title('dataset ve predictions (1st iteration)', fontsize=20)
sns.lineplot(x="x", y="prediction", data=data, color='red', alpha=0.8, ax=ax[0])
ax[0].set_ylim(0,100)
ax[0].set(ylabel='target and prediction')

sns.scatterplot(x="x", y="residual", data=data, cmap='coolwarm', s=100,
                ax=ax[1]).set_title('residuals for target and predictions (1st iteration)', fontsize=20)
sns.lineplot(x="x", y=0, data=data, color='red', alpha=0.5, ax=ax[1])
ax[1].set_ylim(-50,50)

fig.savefig("gb_01_eng.png", dpi=500)
```
